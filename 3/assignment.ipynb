```python
df = pd.read_csv('age2.csv')
yeardata68 = df[['1968', 'status']].groupby('status').sum().reset_index()
yeardata70 = df[['1970', 'status']].groupby('status').sum().reset_index()
yeardata00 = df[['2000', 'status']].groupby('status').sum().reset_index()
yeardata20 = df[['2020', 'status']].groupby('status').sum().reset_index()

yeardata20
merged_data = yeardata68.merge(yeardata70, on='status', how='inner', suffixes=('_1968', '_1970')) \
                        .merge(yeardata00, on='status', how='inner') \
                        .merge(yeardata20, on='status', how='inner', suffixes=('_2000', '_2020'))


merged_data.set_index('status', inplace=True)
merged_data.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Status')
plt.ylabel('Values')
plt.title('Comparison of Values by Status for Different Years')
plt.show()
```

```python
df = pd.read_csv('age2.csv')
yeardata68 = df[['1968', 'status']].groupby('status').sum().reset_index()
yeardata70 = df[['1970', 'status']].groupby('status').sum().reset_index()
yeardata00 = df[['2000', 'status']].groupby('status').sum().reset_index()
yeardata20 = df[['2020', 'status']].groupby('status').sum().reset_index()

yeardata20
merged_data = yeardata68.merge(yeardata70, on='status', how='inner', suffixes=('_1968', '_1970')) \
                        .merge(yeardata00, on='status', how='inner') \
                        .merge(yeardata20, on='status', how='inner', suffixes=('_2000', '_2020'))


merged_data.set_index('status', inplace=True)

merged_data['1968'] = (merged_data['1968'] / merged_data['1968'].sum()) * 100
merged_data['1970'] = (merged_data['1970'] / merged_data['1970'].sum()) * 100
merged_data['2000'] = (merged_data['2000'] / merged_data['2000'].sum()) * 100
merged_data['2020'] = (merged_data['2020'] / merged_data['2020'].sum()) * 100


merged_data.plot(kind='bar', figsize=(10, 6))

plt.xlabel('Status')
plt.ylabel('% Of Total')
plt.title('Comparison of Values by Status for Different Years')
plt.show()
```

```python
df = pd.read_csv('age2.csv')
married_age = df[df['status'] == 'married']
age_columns = ['1968', '1970', '1972', '1974']

plt.figure(figsize=(10, 6))
for year in age_columns:
    plt.plot(married_age['age'], married_age[year], label=year)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Married Couples by Age')
plt.legend(title='Year')
plt.show()
```
